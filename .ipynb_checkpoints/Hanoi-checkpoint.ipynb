### Ward clustering

Now suppose that the investor has opened a first restaurant/shop in the urban area of Hanoi. It was very successful and he wants to find other locations to expand the business. With Foursquare API and utilizing clustering technique, we can help him to find similar location to the current one in terms of popular venues around the location.

```python
# Define a function to get nearby venues of given coordinates using Foursquare API
LIMIT = 100
def getNearbyVenues(names, latitudes, longitudes, radius):
    
    venues_list=[]
    for name, lat, lng, radius in zip(names, latitudes, longitudes, radius):
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

The difference in areas between the largest and smallest wards in the urban area of Hanoi is very big - 124 times. As mentioned above, using fixed radius for Foursquare API is not a  good idea since if the radius is too small, Foursquare might find very few or even not find any venue at  all near a given coordinate in large wards; while if the radius is too big, the nearby venues of a location will overlap with those of its neighbors, especially in the old quarter of Hanoi. So, let define searching radius to each ward.

```python
# Assign radius to use with Foursquare API depending on Wards' areas
# This function make use of the colormap defined above
def getRadius(quant, area):
    if area <= quant.index[0]:
        return 300
    if area <= quant.index[1]:
        return 500
    if area <= quant.index[2]:
        return 800
    if area <= quant.index[3]:
        return 1500
    else: return 2000
```

```python
uhanoi_gdf['radius'] = uhanoi_gdf['Area'].apply(lambda x: getRadius(colormap, x))
uhanoi_gdf[0:10]
```

```python
# Get nearby venues of all the ward centroids
uhanoi_venues = getNearbyVenues(names=uhanoi_gdf['Ward'],
                                   latitudes=uhanoi_gdf['Latitude'],
                                   longitudes=uhanoi_gdf['Longitude'],
                                   radius=uhanoi_gdf['radius']
                                  )
print ('Searching done.')
```

```python
print(uhanoi_venues.shape)
uhanoi_venues[0:10]
```

```python
# Find out how many unique categories can be curated from all the returned venues
print('There are {} unique categories.'.format(len(uhanoi_venues['Venue Category'].unique())))
```

```python
# one hot encoding
uhanoi_onehot = pd.get_dummies(uhanoi_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
uhanoi_onehot['Ward'] = uhanoi_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [uhanoi_onehot.columns[-1]] + list(uhanoi_onehot.columns[:-1])
uhanoi_onehot = uhanoi_onehot[fixed_columns]

uhanoi_onehot.head()
```

```python
#  Next, let group rows by ward and by taking the mean of the frequency of occurrence of each category
uhanoi_grouped = uhanoi_onehot.groupby('Ward').mean().reset_index()
print(uhanoi_grouped.shape)
uhanoi_grouped.head()
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Ward']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Ward'] = uhanoi_grouped['Ward']

for ind in np.arange(uhanoi_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(uhanoi_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
# set number of clusters
kclusters = 3

uhanoi_grouped_clustering = uhanoi_grouped.drop('Ward', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(uhanoi_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster', kmeans.labels_)
```

```python
uhanoi_merged = uhanoi_gdf

# merge uhanoi_gdf with venues data to add latitude/longitude for each ward
uhanoi_merged = uhanoi_merged.join(neighborhoods_venues_sorted.set_index('Ward'), on='Ward')

# Set 99 to the missing values (ward that has no common venues in the given radius from the centroid)
uhanoi_merged.fillna(99, inplace=True)
uhanoi_merged['Cluster'] = uhanoi_merged['Cluster'].astype(int)
uhanoi_merged.head()
```

To give good labels to the clusters, we now make bar chart of number of the most common venues in each cluster

```python
from collections import Counter
def plot_bar(venues, cluster, big=1):
    plt.figure(figsize=(9,6), dpi = 100) if big else plt.figure(figsize=(9,1), dpi = 100)
    venues = Counter(venues)
    venues_df = pd.DataFrame.from_dict(venues, orient='index').reset_index()
    venues_df.columns = ['Venue', 'Count']
    venues_df.sort_values(by='Count', ascending=False, inplace=True)
    venue = venues_df['Venue']
    count = venues_df['Count']
    index = np.arange(len(venue))
    plt.barh(venue, count)
    plt.xlabel('Number of venues', fontsize=8)
    plt.yticks(index, venue, fontsize=8)
    plt.title('Number of most common venues in cluster ' + str(cluster), fontsize=10)
    plt.show()
```

```python
# Let display the clusters on the map
map_clusters = folium.Map(geo_latlon('Trung Liet, Dong Da, Hanoi, Vietnam'), zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Plot the district population density layer
folium.GeoJson(hanoi_gdf.loc[hanoi_gdf['ENGTYPE_2']=='Urban District'],
               ## controls  the fill of the geo regions; applying getColor based on population density
               style_function = lambda x: {'weight': 1, 'color': 'grey',
                                           'fillColor': getColor(x['properties']['Population_density']),
                                           'fillOpacity': 0.7},
              ).add_to(map_clusters)

# Wards
# Hover over for ward and cluster label info
folium.GeoJson(uhanoi_merged,
               ## controls  the fill of the geo regions; applying getColor based on population density
               style_function = lambda x: {'weight': 1, 'color': 'grey',
                                           'fillColor': '#FFFFCC', 
                                           'fillOpacity': 0.3},
               ## changes styling of geo regions upon hover
               highlight_function=lambda x: {'weight':2},
                ## popup Area, District type, population and density information
               tooltip=folium.GeoJsonTooltip(fields=['District', 'Ward', 'Cluster', 'Labels'])
              ).add_to(map_clusters)

# add centers to the map
# Hover over for venues info
for lat, lon, cluster, ward, ven1, ven2, ven3 in zip(uhanoi_merged['Latitude'], 
                                         uhanoi_merged['Longitude'], uhanoi_merged['Cluster'], uhanoi_merged['Ward'], 
                                         uhanoi_merged['1st Most Common Venue'],
                                         uhanoi_merged['2nd Most Common Venue'],
                                         uhanoi_merged['3rd Most Common Venue']):
    color = 'black' if cluster == 99 else rainbow[cluster-1]
    label = folium.Tooltip('<strong>Ward</strong>&nbsp;' + str(ward) + '<br>' + '<strong>Venues</strong>&nbsp;' + \
                           unidecode.unidecode(str(ven1) + ', ' + str(ven2) + ', ' + str(ven3)))
    folium.CircleMarker(
        [lat, lon],
        radius=3,
        color=color,
        tooltip=label,
        fill=True,
        fill_color=color,
        fill_opacity=0.7).add_to(map_clusters)

# show the old quarter area
folium.CircleMarker(
        geo_latlon('Hang Dao, Hoan Kiem, Hanoi, Vietnam'),
        radius=30,
        color='white',
        fill=False).add_to(map_clusters)    


# Add legends
clusters = uhanoi_merged[['Cluster', 'Labels']].drop_duplicates().reset_index(drop=True)

legends = ""
for i in range (0, clusters.shape[0]):
    color = 'black' if clusters.loc[i]["Cluster"] == 99 else rainbow[clusters.loc[i]["Cluster"] - 1]
    legends = legends + '<i style="background-color:' + color + '; ' + \
            'width: 18px; height: 18px; float: left; margin-left: 8px; margin-right: 8px; opacity: 0.7"></i> '  + \
            unidecode.unidecode(clusters.loc[i]['Labels']) + '<br>' 
    
legend_html = '''
                <div style="position: absolute; border-radius: 5px;
                            bottom: 15px; right: 10px; width: 265px; height: 95px; 
                            border: 1px solid grey; z-index:9999; font-size: 12px; 
                            line-height: 18px; color: #555; background-color: white;
                            ">&nbsp;&nbsp;&nbsp;<strong>Clusters</strong><br>''' + legends + ''' 
                </div>
                ''' 
map_clusters.get_root().html.add_child(folium.Element(legend_html))

map_clusters
```

The map clearly shows two dominant clusters - purple and red ones. The purples are places with high-end restaurants, hotels and coffee shops while the red ones have many Fast food and Café. Most of the purples are in in the center of Hanoi (close to the old quarter - the white circle in the map) where the population is very densed, where many visitors come and stay and where there are many office buildings and shopping streets  located. The red ones are located further to the west and south from the center, to the direction that the city has been expanded during the last few years.

The third cluster (light greens) include wards at the edge of the urban area. Those wards are less densed, close to the rivers and suitable for outdoor events. 

There is one ward shown at the bottom left of the map - Dong Mai - with the center colored in black as Foursquare returned no common venue for it, even with the searching radius set to 2km. This shows that choosing appropriate value of searching radius is very important for the analysis.
